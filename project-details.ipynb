## Markov Chain Monte Carlo
solution: helps with complex integrals or sample spaces with hig dimensionality
purpose: draw samples from a Pr() distro
use: use markov chain equilibrium distro to match target distro
how: 
1. pick init state sample
2. generate candidate next tate from distro
3. calculate an acceptance Pr() 
4. accept/reject
5. repeat until equilibrium
6. discard rest
